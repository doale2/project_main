# 2. Project information
    
Wafer-scale data processing refers to a technology that efficiently manages and utilizes vast amounts of data of elements in
each die at the wafer scale to improve the manufacturing process and enhance product quality.
This requires the use of appropriate data formats and processing tools.
    
So our team created a module that can efficiently extract  and analyze the XMLfiles provided by the customer, 
it also shows and saves information in graphs and csv files.  
    
Customers can select and analyze files, lots, wafers, column&row which they want.

# 4. Description of the module file feature
<i class="fa fa-folder" style="font-size: 24px; color: #FFBB00;"></i> .src

- analyze_data.py: The code analyzes the XML file to visualize the data and store the results. You can choose different analysis methods for a given XML file, and generate graphs according to the selected analysis method and save the results. It also updates the progress bar indicating the progress of the analysis and performs error handling.
<br/>

- choose_analysis_scale.py: The code performs the function of navigating XML files in the "./dat" directory and generating a selection list based on information extracted from the file name. Deduplication is generated based on the information "lot", "wafer", "coordinate", and "date" extracted from the file name. Afterwards, the generated selection list is displayed in the GUI.
<br/>

- init_GUI.py: The code given serves to create and initialize the GUI using tkinter. The GUI has various components, including logos, check boxes, options menus, tab views, progress bars, list boxes, and buttons. This code sets the layout and initial values of these elements. Also, call the choose_analysis_scale function to generate a selection list and display it in the GUI.
<br/>

- iv_fitting.py: The given code performs the function of fitting the voltage-current curve for the given IV data. The code first divides the data based on the threshold voltage and applies different fitting methods to the previous and subsequent data. Previous data use polynomial approximation, and subsequent data perform fitting using the model function defined. Finally, the two fitting results are combined into one list and returned.
<br/>

- iv_graph.py: The code parses IV data from XML files and uses matplotlib to visualize the data. In addition, IV fitting is performed on the IV data to draw the optimal curve and save the graph as a PNG file.
<br/>

- main.py: The code initializes the GUI application and handles events that change the appearance mode and sizing based on user input. It also includes methods that perform data analysis functions, store and display results, and methods that terminate applications.
<br/>

- save_choosed_data.py: Finds the file path based on the index you select and saves the path of the selected files. The function displays a list of selected files in a text box.
<br/>

- save_csv.py: Extracts the required data from the XML file and returns it to the list. Extract data from the XML file you extract, and utilize the extracted data to create a DataFrame and save it as a CSV file.
<br/>

- setting_subplots.py: Creates a 3x3 subplot and returns the created subplot. Each subplot sets a title, label, tick, etc. according to the given format Process detailed settings for the subplot. For each subplot, set the title, label, tick size, and so on, and add legends and grids.
<br/>

- ts_fitting.py: The code processes optical transmission data to fit the optimal model and generate graphs. Key features include conversion between dBm and linear values, n_eff extraction from optical transmissions, and fitting a voltage-consideration model.
<br/>

- ts_graph.py: This code parses data from XML files to create optical transmission graphs and provides the ability to fit the optimal model. It also has the ability to calculate RÂ² values by comparing measured data with optimal models, and to perform first-order approximation by extracting peaks.

# 5. Example of analysis result
Customers can extract and analyze XML files provided by the customer using our module, which efficiently handles the data processing. The program also presents the information in the form of graphs and saves it in both **CSV file** and **PNG file** formats.

The CSV files contains comprehensive information, including details such as lot, wafer, mask, testsite, name, date, script ID,
script version, script owner, operator, row, column, error flag, error description, analysis wavelength,
Rsq of the Nth reference spectrum, maximum transmission of the reference spectrum (in dB), Rsq of the IV curve,
and current values at -1V and 1V.

The PNG files include various graphs, such as the measured transmission spectra, flat transmission spectra, IV analysis,
flat flat transmission, n-V curve, and enlarged flat flat transmission spectra (in dBm).
These graphs provide visual representations of the component's performance and characteristics.
