## 2. Project information
    Wafer-scale data processing refers to a technology that efficiently manages and utilizes vast amounts of data of elements in
    each die at the wafer scale. This requires the use of appropriate data formats and processing tools.  
    
    So our team created a module that can efficiently extract  and analyze the XMLfiles provided by the customer, 
    it also shows and saves information in graphs and csv files.  
    
    Customers can select and analyze files, lots, wafers, column&row which they want.
